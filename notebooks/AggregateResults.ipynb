```python
import pandas as pd
```

```python
def read_directed():
    df = pd.read_csv("directed_graphs.csv").drop("Unnamed: 0", axis=1)
    df["persistence"]/=df["num_edges"]
    df["triangles"]/=(df["num_nodes"]*df["num_times"]*3)
    return df

def read_undirected():
    df = pd.read_csv("undirected_graphs.csv").drop("Unnamed: 0", axis=1)
    df["persistence"]/=df["num_edges"]
    df["triangles"]/=(df["num_nodes"]*df["num_times"]*6)
    return df

df = pd.concat([read_directed(), read_undirected()])
```

```python
import numpy as np
```

```python


def get_crosstab(df2, key):
    out = []
    default_digits=3
    for m, s in zip(df2[key]["mean"], df2[key]["std"]):
        if np.isnan(s) or s < 10.0**-10:
            out.append(str(np.round(m,default_digits)))
            continue

        digits = 0
        if s > 10.0**(-10):
            while 10.0**digits * s < 1:
                #print(10.0**digits * s, digits)
                digits += 1
        if digits == 0:
            mean = str(int(np.round(m, digits)))
        else:
            mean = str(np.round(m, digits))
        out.append(mean + "("+str(int(np.round(10.0**digits * s))) +")")
    df_tilt = pd.DataFrame()
    df_tilt["abbr"] = df2["abbr"]
    df_tilt["depth"] = df2["depth"]
    tmp = df_tilt["depth"].copy()
    tmp[tmp>=2]=2
    df_tilt["depth"] = tmp
    df_tilt[key] = out
    df= df_tilt.pivot(index="abbr", columns="depth", values=key).reset_index()
    return df[["abbr", -1, 2, 1, 0]]
```
