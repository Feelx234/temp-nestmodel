```python
edges0 = np.array([[2,1]], dtype=np.uint32)
edges1 = np.array([[1,2]], dtype=np.uint32)
edges2 = np.array([[2,3]], dtype=np.uint32)
edges3 = np.array([[0,1],[1,2], [2,0]], dtype=np.uint32)
G = SparseTempFastGraph([edges0.copy(), edges1.copy(), edges2, edges3, edges0], is_directed=True)
G2 = TempFastGraph([edges0.copy(), edges1.copy(), edges2, edges3, edges0], is_directed=True)
```

```python
draw_networkx_temp(G)
```

```python
draw_networkx_temp(G2, colors = 1)
```
